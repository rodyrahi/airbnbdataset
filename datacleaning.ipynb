```python
# Import liberies
import pandas as pd
```

```python

# Load the Excel file
file_path = './listings2023.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to inspect the data
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Check for duplicates
duplicates = data.duplicated().sum()

# Display basic info about the dataframe
data.info()

missing_values, duplicates
```

```python

data.fillna('N/A', inplace=True)

```

```python
# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert all text to lowercase
data = data.applymap(lambda s: s.lower() if type(s) == str else s)

# Standardize date formats if any date columns exist
# data['date_column'] = pd.to_datetime(data['date_column'], format='%Y-%m-%d')
```

```python

data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Check for duplicates
duplicates = data.duplicated().sum()

# Display basic info about the dataframe
data_info = data.info()

missing_values, duplicates, data_info
```

```python
# Fill missing values in the 'license' column with 'N/A'
data['license'].fillna('N/A', inplace=True)
```

```python
# Convert all text columns to lowercase
text_columns = ['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type', 'license']
for col in text_columns:
    data[col] = data[col].str.lower()
```

```python
# Fill missing values in the 'license' column with 'N/A'
data['license'].fillna('N/A', inplace=True)

# Convert all text columns to lowercase
text_columns = ['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type', 'license']
for col in text_columns:
    data[col] = data[col].str.lower()

# Save the cleaned data to a new file
cleaned_file_path = './listings2023_cleaned.xlsx'
data.to_excel(cleaned_file_path, index=False)

cleaned_file_path
```
